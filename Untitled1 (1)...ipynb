```python

import numpy as np
import scipy.sparse as sp

data = np.genfromtxt('UMI_counts.txt', delimiter='\t', dtype=int, skip_header=1)


barcodes = data[:, 0]
genes = data[:, 1]

barcode_counts = {}
for barcode in np.unique(barcodes):
    barcode_counts[barcode] = np.zeros(data.shape[1] - 1, dtype=int)

for i in range(data.shape[0]):
    barcode = barcodes[i]
    gene = genes[i]
    count = data[i, 2]
    barcode_counts[barcode][gene] = count

barcode_matrix = sp.csr_matrix(list(barcode_counts.values()))

sp.save_npz('barcode_matrix.npz', barcode_matrix)
```
